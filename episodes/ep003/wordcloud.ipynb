```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud
```

```python
df = pd.read_csv("./data/dune_words.csv", index_col=0)
df = df.iloc[1:, ]
df.dropna(axis='rows', inplace=True)
df.head()
```

```python
colors = ['ef3e36', '00bc68', '0061ea', 'f58300', 'f5c206', '05c5d7', 'a459cf']
colors = ['#' + c for c in colors]
    
col_f = lambda *a, **kw: np.random.choice(colors, 1)[0]
```

```python
nw2keep = 300

np.random.seed(1)
wc = WordCloud(
    width = 2000,
    height = 2000,
    max_words=nw2keep,
    background_color='black', 
    color_func=col_f, 
    random_state=42
)

wc.generate_from_frequencies(dict(zip(df['word'], df['freq'])))
fig = plt.figure(figsize=(7, 7))
plt.imshow(wc)
plt.axis('off')
```
